## 3. ***Data Wrangling***

### Data Wrangling Code

**Data Cleaning**

````python
# Write your code to make your dataset analysis ready.

copyed_data = df.copy()                         #lets make a copy of our original data. So that their is no any changes in our original data
copyed_data.drop_duplicates(inplace = True)              # drop duplicate rows present in our data
````

````python
# delete company column from our dataset(because it contain lakhs of null values)
# use try and except to handel the errors
try:
  copyed_data.drop('company', axis = 1, inplace = True)
except KeyError:
  print('KeyError: Somthing went wrong.\nHint run from the above from where you load your dataset')
````

````python
# Replace all null rows of agent, country, children with 0,missing_country_code,0
# we should not drop as much as possible because it lead to loose data. So here we use fillna method to replace null rows with others values
copyed_data['agent'].fillna(0,inplace = True)
copyed_data['country'].fillna('missing_country_code',inplace = True)
copyed_data['children'].fillna(0,inplace = True)
````

````python
# replace all values of meal with sc where it is undefined(because it is similar to sc)

copyed_data['meal'].replace({'Undefined':'SC'}, inplace = True)

#replace 1 with repeated guest and 0 with not repeated guest in is_repested_guest column
copyed_data['is_repeated_guest'].replace({1:'repeated guest',0:'not repeated guest'},inplace = True)
````

````python
# Lets create some more usefull columns which is helpfull for getting insights

copyed_data['total_stay_in_nights'] = copyed_data['stays_in_week_nights'] + copyed_data['stays_in_weekend_nights']   # create a column of total stay in nights
copyed_data['revenue'] = copyed_data['adr'] * copyed_data['total_stay_in_nights']                                      # create a column of revenue
````

````python
copyed_data.head()
````

### What all manipulations have you done and insights you found?

Till now we will copy the original data and delete all the duplicate rows and drop company column which contain 1,12,593 null values. We replace null values of columns of agent and children with 0 or country column  with missing country code. After that we will replace the Undefined with SC in meal column.

We will create two columns firts is total stay in nights which we will getting from adding of two columns stays in week nights and stays in weekend nights and second is revenue column in which we multiplying two columns adr(average daily rates) and total stay in nights. From revenue column we find the maximum revenue on the basis of month corresding to the hotel type.

What is the revenue trend by month on the basis of hotel ?

````python
#make a list of hotel type
hotel_type = sum_of_revenue['hotel'].unique()

#here we plot a line chart using matplotlib
plt.figure(figsize = (12,5))
plt.plot(city_hotel_data['arrival_date_month'],city_hotel_data['revenue'],color = 'tomato',marker = "o",markersize = 10)
plt.plot(resort_hotel_data['arrival_date_month'],resort_hotel_data['revenue'],color = 'mediumpurple',marker = "*",markersize = 10)
plt.legend(hotel_type)
plt.title('Revenue trend by month',bbox = {'facecolor':'mintcream','boxstyle':'round'})
plt.xlabel('month',bbox = {'facecolor':'ghostwhite','boxstyle':'round'})
plt.ylabel('sum of revenue',bbox = {'facecolor':'ghostwhite','boxstyle':'round'})
plt.show()
````

````python
# Chart - 7 visualization code
#get the data of hotel country count
hotel_country_count = copyed_data.groupby('hotel')['country'].value_counts(ascending = False).reset_index()

# get the data of first 5 rows of of city and resort hotel
City_Hotel = hotel_country_count[hotel_country_count['hotel']=='City Hotel'].head()
Resort_Hotel = hotel_country_count[hotel_country_count['hotel']=='Resort Hotel'].head()
'''
we know that in resort hotel their is a country column in which their is IRL which is not preset in City_Hotel
So here we find the data of DEU present in country of resort hotel type
'''
DEU_country = hotel_country_count[(hotel_country_count['hotel']== 'Resort Hotel') & (hotel_country_count['country']=='DEU')]

# using numpy we concat DEU_country in Resort_Hotel
Resort_Hotel = pd.concat([Resort_Hotel,DEU_country],axis = 0) #argument are: np.append(array,value,axis)

# we concat city and resort hotel here
concat = pd.concat([Resort_Hotel, City_Hotel],axis = 0)
# we drop row where irl is present (it is because in top 5 city hotel there is no IRL country so we but it is present in resort hotel)
concat.drop(170,inplace = True)

concat
````

````python
# convert row to column and column to row using pivot
#this is a group bar chart
concat.pivot(index = 'hotel',columns ='country',values = 'count').plot(kind = 'bar',figsize = (10,7),edgecolor = 'black')
plt.title('Top 5 country by hotel',bbox = {'facecolor':'azure'})
plt.xlabel('Hotel type',bbox = {'facecolor':'azure'})
plt.ylabel('Number of bookings',bbox = {'facecolor':'azure'})
plt.legend(loc = 'upper left')
plt.show()
````
